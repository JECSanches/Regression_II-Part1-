```python
# Carregando bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Carregando dataset
df = pd.read_csv('previsao_de_renda.csv')
print(df.shape)
df.head(3)
```

# 1. 

```python
# Removendo colunas desnecessárias à análise
df = df.drop(['Unnamed: 0', 'data_ref', 'id_cliente'], axis=1)
df.shape
```

```python
# Removendo dados faltantes
print(df.isnull().sum())
```

```python
df = df.dropna()
print(df.shape)
df.isna().sum()
```

```python
# Variáveis qualitativas

df.select_dtypes('object').describe().transpose()
```

#### <span style='color:blue'> Gerando o modelo

```python
# Gerando matriz design e y

y, x = patsy.dmatrices('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda)' 
                       '+ C(educacao) + C(estado_civil) + C(tipo_residencia) + idade + tempo_emprego + qt_pessoas_residencia',
                       data = df)

reg = sm.OLS(y, x).fit()
reg.summary()
```

# 2.

Para o modelo em questão, a variável menos significativa é "tipo_residencia". Vamos removê-lá.

```python
# Removendo a variável menos significante "tipo_residencia"

reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda)' 
                       '+ C(educacao) + C(estado_civil) + idade + tempo_emprego + qt_pessoas_residencia',
                       data = df).fit()
reg.summary()
```

<span style='color:purple'>**Note que ao remover a variável "tipo_residencia" os valores de $R^{2}$ e `R-quadrado ajustado` nem mesmo foram alterados. Já o valor de `AIC` diminuiu um pouco ($[2.719e+04] \rightarrow [2.718e+04]$), o que indica uma melhora do modelo.**

# 3.

Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 


- De acordo com a análise, podemos remover as variáveis: `educacao`, `estado_civil` e `qt_pessoas_residencia`. A variável `qt_pessoas_residencia` apesar de possuir um bom p-value possui um intervalo de confiança muito grande, dessa forma sua remoção não afetará o modelo.
- A variável `tipo_renda` possui alguns valores de p-value e intervalos de confiança questionáveis (`C(tipo_renda)[T.Bolsista]` e `C(tipo_renda)[T.Pensionista]`), mas também possui "subvariáveis" com bons valores de p-value e intervalos de confiança (`C(tipo_renda)[T.Empresário]` e `C(tipo_renda)[T.Servidor público]`), logo será mantida no modelo. 

```python
reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda)' 
                       '+ idade + tempo_emprego',
                       data = df).fit()

reg.summary()
```

#### Insights

- Comparando os modelos podemos concluir que após a remoção das variáveis menos significativas, os valores de $R^{2}$ e *Adj. R-quadrado* diminuíram indicando que o modelo piorou em qualidade.
- O valor associado ao parâmetro `AIC` aumentou e isso indica uma piora na qualidade do modelo.
- Um ponto a se destacar é que o modelo final, apesar de possuir indicadores um pouco piores, é capaz de explicar bem a variabilidade dos dados, se comparado ao primeiro modelo (`R-quadrado` partiu de 0.357, no primeiro modelo, para 0.353, no último). Logo, o modelo final tem bons valores e não está sobrecarregado de variáveis pouco explicativas, evitando assim um possível overfitting.
